**Задача 2.1:** Данные о выручке сервиса представлены в российских рублях и казахстанских тенге. Приведите выручку к единой валюте — российскому рублю.

Для этого используйте датасет с информацией о курсе казахстанского тенге по отношению к российскому рублю за 2024 год — `final_tickets_tenge_df.csv`. Его можно загрузить по пути `https://code.s3.yandex.net/datasets/final_tickets_tenge_df.csv')`

Значения в рублях представлено для 100 тенге.

Результаты преобразования сохраните в новый столбец `revenue_rub`.

---


```python
# загружаем файл с курсом валют
exchange_rate_df = pd.read_csv('https://code.s3.yandex.net/datasets/final_tickets_tenge_df.csv')
```

```python
# смотрим из чего он состоит
exchange_rate_df.info()
```

```python
# видим, что в зависимости от даты различный курс, поэтому нам надо переводить тенге в рубли с учетом даты. 
# Первый шаг - перевод в формат даты загруженного файла, столбца дата
exchange_rate_df['data'] = pd.to_datetime(exchange_rate_df['data'])
```

```python
# Теперь присоединяем загруженный файл к основной таблице. Чтобы это сделать используем функцию merge. 
# Так как наиболее полные данные в основной таблице, то присоединение по левому принципу
# Даты курса присоединяем к датам заказа
df_with_rate = df.merge(exchange_rate_df, left_on='order_dt', right_on='data', how='left')
```

```python
# присоединение проведено. Теперь проверяем столбец с информацией о типе валюты и если рубль, то оставляем значение как есть.
# но если тенге, то переводим с учетом присоединенного столбца курса.
def convert_to_rub(row):
    if row['currency_code'] == 'rub':
        return row['revenue']  # уже в рублях — оставляем как есть
    elif row['currency_code'] == 'kzt' and pd.notna(row['curs']):
        # Курс дан на 100 тенге → делим revenue на 100 и умножаем на курс
        return (row['revenue'] / 100) * row['curs']
    else:
        return None
```

```python
# далее вставляем нужный по заданию столбец
df_with_rate['revenue_rub'] = df_with_rate.apply(convert_to_rub, axis=1)
```

```python
# и удаляем лишние
result_df = df_with_rate.drop(columns=['curs', 'data', 'nominal', 'cdx'])
```

### 3. Создание профиля пользователя

**Задача 3.1.** Постройте профиль пользователя — для каждого пользователя найдите:

- дату первого и последнего заказа;
- устройство, с которого был сделан первый заказ;
- регион, в котором был сделан первый заказ;
- билетного партнёра, к которому обращались при первом заказе;
- жанр первого посещённого мероприятия (используйте поле `event_type_main`);
- общее количество заказов;
- средняя выручка с одного заказа в рублях;
- среднее количество билетов в заказе;
- среднее время между заказами.

После этого добавьте два бинарных признака:

- `is_two` — совершил ли пользователь 2 и более заказа;
- `is_five` — совершил ли пользователь 5 и более заказов.

**Рекомендация:** перед тем как строить профиль, отсортируйте данные по времени совершения заказа.

---


```python
# начнем пошагово. Сначала отфильтруем данные по времени заказа. 
filtered_two_df = filtered_two_df.sort_values('order_ts').reset_index(drop=True)
filtered_two_df.head()
```

```python
# Теперь надо сгруппировать данные по пользователям (user_id) и посчитать по ним статистику
# Напомним, что при разведке данных было определено 22 000 уникальных пользователя
# общее количество заказов;
# средняя выручка с одного заказа в рублях;
# среднее количество билетов в заказе;
# дату первого и последнего заказа;
# среднее время между заказами.
user_stats = filtered_two_df.groupby('user_id').agg(
    total_orders=('order_id', 'count'),  # общее кол-во заказов
    total_revenue=('revenue_rub', 'sum'),  # суммарная выручка
    total_tickets=('tickets_count', 'sum'),  # суммарное кол-во билетов
    mean_revenue=('revenue_rub', 'mean'), # средняя выручка
    mean_tickets=('tickets_count', 'mean'), # среднее количество билетов в заказе
    first_order_date=('order_ts', 'min'),  # дата первого заказа
    last_order_date=('order_ts', 'max')   # дата последнего заказа
).reset_index()
```

```python
# для работы с пропусками понадобится еще одна библиотека 
import numpy as np
```

```python
# Среднее время между заказами (в днях)
user_stats['avg_days_between_orders'] = (
    (user_stats['last_order_date'] - user_stats['first_order_date']) /
    pd.to_timedelta(1, unit='D') /  # преобразуем в число дней
    (user_stats['total_orders'] - 1)  # делим на (N-1) интервалов между N заказами
)
# Если у пользователя 1 заказ, среднее время = NaN
user_stats['avg_days_between_orders'] = user_stats['avg_days_between_orders'].replace([np.inf, -np.inf], np.nan)
```

```python
user_stats
```

```python
# находим бинарные признаки
user_stats['is_two'] = (user_stats['total_orders'] >= 2).astype(int)
user_stats['is_five'] = (user_stats['total_orders'] >= 5).astype(int)
```

```python
user_stats
```

```python
# делаем таблицу только по первым заказам
first_order = filtered_two_df.groupby('user_id').first().reset_index()
first_order
```

```python
# Добавляем параметры для первых заказов
# устройство, с которого был сделан первый заказ;
# регион, в котором был сделан первый заказ;
# билетного партнёра, к которому обращались при первом заказе;
# жанр первого посещённого мероприятия (используйте поле event_type_main);
user_profile = user_stats.merge(
    first_order[['user_id', 'device_type_canonical', 'region_name', 'service_name', 'event_type_main']],on='user_id',how='left')
user_profile
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(data=user_profile_filter, x='mean_tickets', bins=20, kde=True, alpha = 0.3, stat='density', common_norm=False)

# Подписываем и отображаем график
plt.title('Распределение данных в столбце mean_tickets')
plt.ylabel('Количество')
plt.show()
```

Полимодальное распределение. Выявлено четыре пика, что подтверждает необходимость кластеризации (сегментации) на четыре группы

```python
# проведем сегментацию пользователей:
def assign_ticket_segment(mean_tickets):
    if 1 <= mean_tickets < 2:
        return '1–2 билета'
    elif 2 <= mean_tickets < 3:
        return '2–3 билета'
    elif 3 <= mean_tickets < 5:
        return '3–5 билетов'
    else:  # 5 и более
        return '5+ билетов'
user_profile_filter = user_profile_filter.copy()
user_profile_filter['ticket_segment'] = user_profile_filter['mean_tickets'].apply(assign_ticket_segment)
user_profile_filter.head()
```

```python
# подсчет пользователей
df_count = user_profile_filter.groupby('ticket_segment')['user_id'].count()
df_counts = user_profile_filter.groupby('ticket_segment')['is_two'].value_counts().unstack(fill_value=0)
df_percent = (df_counts.div(df_counts.sum(axis=1), axis=0) * 100).round(2)
df_count
```

Количество билетов в группах 2-3 и 3-5 практически одинаковое, а значит их можно сопоставлять. В категории 1-2 пользователей в 3,5 раза меньше. Людей, купивших более 5 совсем мало, всего 3% от общего числа.

```python
df_counts
```

```python
df_percent
```

Явно выделяется категория людей, купивших 2-3 билета, процент повторно купивших билет составляет 73%, что намного выше сопостовимой категории людей с 3-5 билетами. Получается, что если пользователь купил два или три билета, то скорее он купит еще, но до тех пор, пока не накупит 5 билетов.

#### 4.4. Корреляционный анализ количества покупок и признаков пользователя

Изучите, какие характеристики первого заказа и профиля пользователя могут быть связаны с числом покупок. Для этого используйте универсальный коэффициент корреляции `phi_k`, который позволяет анализировать как числовые, так и категориальные признаки.

---

**Задача 4.4.1:** Проведите корреляционный анализ:
- Рассчитайте коэффициент корреляции `phi_k` между признаками профиля пользователя и числом заказов (`total_orders`). При необходимости используйте параметр `interval_cols` для определения интервальных данных.
- Проанализируйте полученные результаты. Если полученные значения будут близки к нулю, проверьте разброс данных в `total_orders`. Такое возможно, когда в данных преобладает одно значение: в таком случае корреляционный анализ может показать отсутствие связей. Чтобы этого избежать, выделите сегменты пользователей по полю `total_orders`, а затем повторите корреляционный анализ. Выделите такие сегменты:
    - 1 заказ;
    - от 2 до 4 заказов;
    - от 5 и выше.
- Визуализируйте результат корреляции с помощью тепловой карты.
- Ответьте на вопрос: какие признаки наиболее связаны с количеством заказов?

---

```python
import pandas as pd
import phik
from phik import report

# Шаг 1. Подготовка данных
interval_cols = [
    'total_revenue',
    'total_tickets',
    'mean_revenue',
    'mean_tickets',
    'avg_days_between_orders'
]

cols_to_analyze = [
    'total_orders',
    'total_revenue',
    'total_tickets',
    'mean_revenue',
    'mean_tickets',
    'avg_days_between_orders',
    'is_two',
    'is_five',
    'device_type_canonical',
    'region_name',
    'service_name',
    'event_type_main'
]

df_subset = user_profile_filter[cols_to_analyze].copy()

# Шаг 2. Расчёт матрицы Phik
phik_matrix = df_subset.phik_matrix(interval_cols=interval_cols)


# Шаг 3. Извлечение корреляций с total_orders
correlations_with_total_orders = phik_matrix['total_orders'].sort_values(ascending=False)


print(correlations_with_total_orders.round(3))

# Шаг 4. Визуализация (опционально)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(
    phik_matrix,
    annot=True,
    cmap='coolwarm',
    center=0,
    square=True,
    fmt='.2f',
    cbar_kws={'shrink': 0.8}
)
plt.title('Матрица корреляции Phik (φK)', fontsize=14, pad=20)
plt.tight_layout()
plt.show()
```

От общего количества заказов наиболее всего зависят такие признаки как: общее количество билетов, общее количество платежей и среднее количество дней между заказами.

А вот люди, которые совершили покупку повторно, заплатили за первую в среднем 450 - 800 рублей. Различие между группами очевидно: если пользователю достался билет бесплатно, то повторная его покупка маловероятна.
10. Если пользователь купил два или три билета, то с вероятностью 73% он купит еще, но до тех пор, пока не накупит 5 билетов.
11. Наибольшое количество покупок совершено в пятницу и субботу. На повторную покупку день недели не влияет.
12. Среднее время между заказами для клиентов, которые совершили 2-4 заказа составляет 21 день, а для клиентов с билетами более 5 составляет 10 дней. Люди, которые покупают больше билетов и чаще, делают это с меньшим интервалом во времени, нежели люди, совершившие 2-4 заказа.
13. От общего количества заказов наиболее всего зависят такие признаки как: общее количество билетов, общее количество платежей и среднее количество дней между заказами.

**Рекомендации**
Для увеличения возврата клиентов необходимо исключить промокоды и другие стимулирующие действия, которые снижают цену билета. Также советуем предлагать пользователям больше театральных постановок и концертов в мобильных рассылках, так как с телефонов заказывают больше билетов. Рассылки лучше делать накануне пятницы и субботы.

### 6. Финализация проекта и публикация в Git

Когда вы закончите анализировать данные, оформите проект, а затем опубликуйте его.
